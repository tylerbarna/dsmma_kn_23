```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_csv('./ZTF23absbqun.csv')
df.head()
```

```python
df = df[df['instrument_name'] == 'ZTF']
```

```python
df
```

```python
filters = [df[df['filter'] == filter] for filter in df['filter'].unique()]
filters[0]
```

```python
## convert it to a list of lists for each filter, in format [mjd, mag, magerr] for each data point in each filter
data = []
for filter in filters:
    data.append(filter[['mjd', 'mag', 'magerr']].values.tolist())
data
## replace all NaN values with np.inf
data = [[np.nan_to_num(point, nan=np.inf).tolist() for point in filter] for filter in data]
```

```python
df_dict = {}
for filter, data in zip(df['filter'].unique(), data):
    df_dict[filter] = data
df_dict
## replace all NaNs with np.Inf
```

```python
with open('ZTF23absbqun.json', 'w') as fp:
    json.dump(df_dict, fp, indent=4, sort_keys=True)
```
